## Arithmetic Operations

Note: for all of those operations, the images must have the same dimensions, type, and quantity of channels.

##### Library Importing

```python
import cv2 as cv
from matplotlib import pyplot as plt
%matplotlib inline
```

##### Image Reading

```python
img1 = cv.imread("./Resources/Img/Superman.png", 0)
img2 = cv.imread("./Resources/Img/batman.png", 0)

fig = plt.figure(figsize=(9,7))

ax1 = fig.add_subplot(121)
plt.imshow(img1, cmap=plt.cm.gray) 

ax2 = fig.add_subplot(122)
plt.imshow(img2, cmap=plt.cm.gray)

plt.show()
```

### Addition Operation

The addition operation consists of adding the pixels values of both images to each other. With a new image as output.

`add(img1, img2)` <br>

* Input: <br>
    1. `img1` and `img2`: Images to be added; <br><br>

* Output: Matrix with added pixels values.

```python
addedImg = cv.add(img1, img2)

# Added Image Ploting
fig = plt.figure(figsize=(20,5))
plt.imshow(addedImg, cmap=plt.cm.gray)

plt.show()
```

### Subtraction Operation

The subtraction operation consists of subtract the pixels values of both images to each other. With a new image as output.

`subtract(img1, img2)` <br>

* Input: <br>
    1. `img1` and `img2`: Images to be subtracted; <br><br>

* Output: Matrix with subtracted pixels values.

```python
subImg = cv.subtract(img1, img2)

# Added Image Ploting
fig = plt.figure(figsize=(20,5))
plt.imshow(subImg, cmap=plt.cm.gray)

plt.show()
```

### Mixing Operation

The mixing operation consists, with lost of data, of merge the information of two imagens in one.

`addWeighted(src1, alpha, src2, beta, gamma)` <br>

* Input: <br>
    1. `src1`: Matrix referring to the first image; <br>
    2. `alpha`: First image intensity (from 0.0 - transparent to 1 - max intensity); <br>
    3. `src2`: Matrix referring to the second image; <br>
    4. `beta`: Second image intensity (from 0.0 - transparent to 1 - max intensity); <br>
    5. `gamma`: Scalar value added to each sum (can be used to brightness adjustment). <br> <br>

* Output: Matrix with added pixels values.

```python
totalImg = cv.addWeighted(img1, 0.3, img2, 0.3, 0)

# Added Image Ploting
fig = plt.figure(figsize=(20,5))
plt.imshow(totalImg, cmap=plt.cm.gray)

plt.show()
```
