## Histogram Equalization
The purpose of histogram equalization is to modify the tonality of the pixels in the image in order to redistribute the histogram precisely because of the possible bad scattering of the levels, both exposure and contrast. Therefore, a sharper image is obtained containing pixels ranging only from 0 to 255.

##### Library Importing

```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

##### Image Reading
The `imread()` function has a second argument that specifies how the image should be read. They can be:

* `cv.imread_color`: Reads the image as colorized
* `cv.imread_grayscale`: Reads the image in gray scale;
* `cv.imread_unchanged`: Reads the image including the alpha channel.

One can simply pass these arguments with the integers `1`, `0` and `-1` respectively.

```python
img = cv.imread("./Resources/Img/einstein.jpg", 0)
```

##### Equalize Histogram Function

`equalizeHist(img)` <br>
* Input: <br>
    1.	`img`: Image represented as matrix. <br><br>
* Output: New image with equalized histogram.


```python
imgEqualized = cv.equalizeHist(img)
```

##### Gray Scale Image Plot - Original and Equalized

```python
fig = plt.figure(figsize=(9,7))

ax1 = fig.add_subplot(121)
plt.imshow(img, cmap=plt.cm.gray) # cmap=plt.cm.gray to make image into gray scale

ax2 = fig.add_subplot(122)
plt.imshow(imgEqualized, cmap=plt.cm.gray)
plt.show()
```

```python
fig = plt.figure(figsize=(20,5))

ax1 = fig.add_subplot(121)
plt.hist(img.ravel(), 256, [0, 255])

ax2 = fig.add_subplot(122)
plt.hist(imgEqualized.ravel(), 256, [0, 255])
plt.show()
```

## Colorize Image Plot - Original and Equalized

```python
# Image Reading
colorImg = cv.imread("./Resources/Img/dark_city.png", 1)

# Channel Spliting
blue, green, red = cv.split(colorImg)

# Equalized Colorize Image Ploting
colorFig = plt.figure(figsize=(21,21))

# Equalizing the colorize image (each color channel apart)
blueEqualized = cv.equalizeHist(blue)
redEqualized = cv.equalizeHist(red)
greenEqualized = cv.equalizeHist(green)

# Merging each color channel in one img
resultEqualization = cv.merge([blueEqualized, redEqualized, greenEqualized])

# Colorize Image Ploting - Original
ax10 = colorFig.add_subplot(121)
plt.imshow(colorImg)
plt.title("Original Image")

# Colorize Image Ploting - Equalized
ax20 = colorFig.add_subplot(122)
plt.imshow(resultEqualization)
plt.title("Equalized Image")


# Equalized Colorize Image Ploting
# ax20 = colorFig.add_subplot(242)
# plt.imshow(blueEqualized, cmap=plt.cm.Blues)
# plt.title("Equalized Img - Blue Channel")

# ax30 = colorFig.add_subplot(141)
# plt.imshow(greenEqualized, cmap=plt.cm.Greens)
# plt.title("Equalized Img - Green Channel")

# ax40 = colorFig.add_subplot(142)
# plt.imshow(redEqualized, cmap=plt.cm.Reds)
# plt.title("Equalized Img - Red Channel")

plt.show()
```

### Plotting the Histogram Equalized and Original of Color Image

```python
colorFig = plt.figure(figsize=(20,5))

ax1_color = colorFig.add_subplot(121)
ax1_color.hist(colorImg.ravel(), 256, [0,256])
plt.title("Original Colorize Histogram")

ax2_color = colorFig.add_subplot(122)
ax2_color.hist(resultEqualization.ravel(), 256, [0,256])
plt.title("Equalized Colorize Histogram")


# ax1_color = colorFig.add_subplot(133)
# ax1_color.hist(blue.ravel(), 256, [0,256])
# plt.title("Colorize Histogram Channel Blue")

# ax2_color = colorFig.add_subplot(132)
# ax2_color.hist(green.ravel(), 256, [0,256])
# plt.title("Colorize Histogram Channel Green")

# ax3_color = colorFig.add_subplot(133)
# ax3_color.hist(red.ravel(), 256, [0,256])
# plt.title("Colorize Histogram Channel Red")

plt.show()
```
