## Histogram Equalization
The purpose of histogram equalization is to modify the tonality of the pixels in the image in order to redistribute the histogram precisely because of the possible bad scattering of the levels, both exposure and contrast. Therefore, a sharper image is obtained containing pixels ranging only from 0 to 255.

##### Library Importing

```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

##### Image Reading
The `imread()` function has a second argument that specifies how the image should be read. They can be:

* `cv.imread_color`: Reads the image as colorized
* `cv.imread_grayscale`: Reads the image in gray scale;
* `cv.imread_unchanged`: Reads the image including the alpha channel.

One can simply pass these arguments with the integers `1`, `0` and `-1` respectively.

```python
img = cv.imread("./Resources/Img/einstein.jpg", 0)
```

##### Equalize Histogram Function

`equalizeHist(img)` <br>
* Input: <br>
    1.	`img`: Image represented as matrix. <br><br>
* Sa√≠da: New image with equalized histogram.


```python
imgEqualized = cv.equalizeHist(img)
```

##### Gray Scale Image Plot - Original and Equalized

```python
fig = plt.figure(figsize=(9,7))

ax1 = fig.add_subplot(121)
plt.imshow(img, cmap=plt.cm.gray) # cmap=plt.cm.gray to make image into gray scale

ax2 = fig.add_subplot(122)
plt.imshow(imgEqualized, cmap=plt.cm.gray)
plt.show()
```

```python
fig = plt.figure(figsize=(20,5))

ax1 = fig.add_subplot(121)
plt.hist(img.ravel(), 256, [0, 255])

ax2 = fig.add_subplot(122)
plt.hist(imgEqualized.ravel(), 256, [0, 255])
plt.show()
```
