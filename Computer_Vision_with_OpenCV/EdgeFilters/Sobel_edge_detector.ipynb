## Sobel Edge Detector 

The Sobel operation is used to enhance contours in an image. This non-linear filter highlight horizontal and vertical lines. The result is an image with thicker edges.

##### Library Importing

```python
import cv2 as cv
from matplotlib import pyplot as plt
%matplotlib inline
```

##### Image Reading

```python
img = cv.imread('./Resources/Img/building.JPG')
```

#### Sobel Function

`Sobel(img, var_arm, v, h, mask)` <br>
* Input: <br>
    1.	`img`: Image which the filter will be applied; <br>
    2.	`var_arm`: Value representing the pixel; <br>
    3.	`v`: Vertical Highlight (`0` or `1`); <br>
    4.  `h`: Horizontal Highlight (`0` or `1`); <br>
    5.  `mask`: Dimension of the mask that will be applied to the image (`1`, `3`, `5`, `7` or `9`). <br>

* Output: matrix referring to the image with sobel filter enhancement. <br><br>


```python
# CV_8U: pixels will be treated as integer values between 0 and 255 and stored in 8 bytes
sobelX = cv.Sobel(img, cv.CV_8U, 0, 1, ksize=5)
sobelY = cv.Sobel(img, cv.CV_8U, 1, 0, ksize=5)
```

```python
# Image Plotting
fig = plt.figure(figsize=(20, 50))

ax1 = fig.add_subplot(131)
plt.imshow(img)
plt.title("Original Image")

ax2 = fig.add_subplot(132)
plt.imshow(sobelX)
plt.title("Image with Sobel Filter Applied in X")

ax3 = fig.add_subplot(133)
plt.imshow(sobelY)
plt.title("Image with Sobel Filter Applied in Y")
plt.show()
```
