## Laplacian Edge Detector 

The Laplacian filter is more efficient than Sobel to enhance contours in an image. This non-linear filter has a mask of order 3 that runs through the entire image changing the target pixel by the weighted average of its adjacent ones and then squares this value obtained. The result is an image with thinner edges compared to a Sobel filter, however, it is also very sensitive to noises.

##### Library Importing

```python
import cv2 as cv
from matplotlib import pyplot as plt
%matplotlib inline
```

##### Image Reading

```python
img = cv.imread('./Resources/Img/building.JPG')
```

#### Laplacian Function

`Laplacian(img, var_arm)` <br>
* Input: <br>
    1.	`img`: Image which the filter will be applied; <br>
    2.	`var_arm`: Value representing the pixel. <br>
        `CV_8U`: pixels will be treated as integer values (between 0 and 255) and stored in 8 bytes. <br><br>

* Output: matrix referring to the image with sobel filter enhancement. <br><br>

```python
# CV_8U: pixels will be treated as integer values between 0 and 255 and stored in 8 bytes
laplacian = cv.Laplacian(img, cv.CV_8U)
```

```python
# Image Plotting
fig = plt.figure(figsize=(20, 50))

ax1 = fig.add_subplot(121)
plt.imshow(img)
plt.title("Original Image")

ax2 = fig.add_subplot(122)
plt.imshow(laplacian)
plt.title("Image with Laplacian Filter Applied")
plt.show()
```
