## Canny Edge Detector 

The Canny filter essentially pursues three goals:

* Low error rates
* The distance between a point marked "edge" and the center of the true edge should be as small as possible
* Must not identify multiple edge points where there is only a single edge pixel

##### Library Importing

```python
import cv2 as cv
from matplotlib import pyplot as plt
%matplotlib inline
```

##### Image Reading

```python
img = cv.imread('./Resources/Img/building.JPG')
```

#### Canny Function

`Canny(img, detc_int1, detc_int2)` <br>
* Input: <br>
    1.	`img`: Image which the filter will be applied; <br>
    2.	`detc_int1`: Intensity detection 1; <br>
    3.  `detc_int2`: Intensity detection 2. <br><br>

* Output: matrix referring to the image with Canny filter enhancement. <br><br>

```python
canny = cv.Canny(img, 75, 50)
```

```python
# Image Plotting
fig = plt.figure(figsize=(20, 50))

ax1 = fig.add_subplot(121)
plt.imshow(img, cmap=plt.cm.gray)
plt.title("Original Image")

ax2 = fig.add_subplot(122)
plt.imshow(canny, cmap=plt.cm.gray)
plt.title("Image with Canny Filter Applied")
plt.show()
```
