## Cartoon Filter 

Applying cartoon style filters to images

##### Library Importing

```python
import cv2 as cv
from matplotlib import pyplot as plt
%matplotlib inline
```

##### Image Reading

```python
img = cv.imread('./Resources/Img/sunset.png')

plt.figure(figsize=(10,8))
plt.imshow(img)
plt.show()
```

#### Edge Preserving Filter

This technique softens noise or textures while keeping **edges sharp**.

`edgePreservingFilter(img, dest, flags,sigma_s, sigma_r)` <br>
* Input: <br>
    1.	`img`: Image which the filter will be applied; <br>
    2.	`dest` (optional): Output image; <br>
    3.  `flags`: Edge preserving filters; <br>
        `RECURS_FILTER` (Recursive Filter - `flags = 1`): Around 3.5x faster than NORMCONV_FILTER; <br>
        `NORMCONV_FILTER` (Normalized Convolution - `flags = 2`): Makes border sharp and is good to aplications as stylizations; <br>
    4.  `sigma_s`: Controls the size of adjacent for smoothing - values between `0` and `200`; <br>
    5.  `sigma_r`: Controls how different colors in the neighborhood will be calculated - values between `0` and `1`. <br>

* Output: Image with Edge Preserving Filter enhancement. <br><br>

```python
EdgePreservingImg = cv.edgePreservingFilter(img, flags=2, sigma_s=80, sigma_r=0.4)
```

```python
# Image Plotting
fig = plt.figure(figsize=(20, 50))

plt.imshow(EdgePreservingImg)
plt.title("Image with Edge Preserving Filter Applied")
plt.show()
```

#### Stylization Filter

`stylization(img, sigma_s, sigma_r)` <br>
* Input: <br>
    1.	`img`: Image which the filter will be applied; <br>
    2.  `sigma_s`: Controls the size of adjacent for smoothing - values between `0` and `200`; <br>
    3.  `sigma_r`: Controls how different colors in the neighborhood will be calculated - values between `0` and `1`. <br>

* Output: Image with Stylization Filter enhancement applied. <br><br>

```python
cartoonImg = cv.stylization(img, sigma_s=160, sigma_r=0.2)
```

```python
# Image Plotting
fig = plt.figure(figsize=(20, 50))

plt.imshow(cartoonImg)
plt.title("Image with Stylization Filter Applied")
plt.show()
```
