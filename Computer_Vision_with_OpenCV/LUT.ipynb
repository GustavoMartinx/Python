## LUT - Look Up Table

It's a way to apply image transformations quickly and efficiently. The goal is to create a table of pre-calculated values used to map each pixel in the image to a new value. It can be used to apply filters, adjust contrast, change colors, etc.

##### Library Importing

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

##### Image Reading

```python
img = cv.imread('./Resources/Img/screws.jpg')
plt.imshow(img)
plt.show()
```

### The LUT works like this:

**Table Creation**:  The first step is create the Look Up Table; <br>
**Table Aplication**:  After that, it is used to map each pixel in the image to a new value; <br>
**Final Result**:  The output image will look different from the original image depending on the table used. <br>

`LUT(img, Table)`
* Input: <br>
    1.	`img`: Original image; <br>
    2.  `Table`: Look Up Table. <br><br>

* Output: Imagem com a aparncia ajustada ou modificada aps a aplicao da tabela.<br>

### Creating a LUT to Increase the Contrast of an Image

```python
# Creating LUT to Increase Contrast
lut = np.zeros(256, dtype = 'uint8')
for i in range(256):
    if i < 200:
        lut[i] = 0
    else:
        lut[i] = 255

# Applying the LUT on the image
contrastedImg = cv.LUT(img, lut)

# Plotting the Original image and
# the image with Increased Contrast
plt.figure(figsize=(15,8))
plt.subplot(1, 2, 1)
plt.imshow(img)
plt.title('Original Image')

plt.subplot(1, 2, 2)
plt.imshow(contrastedImg)
plt.title('Increased Contrast')
plt.show()
```

### Creating a LUT to Change the Image Colors

```python
# Creating a LUT to Change the Image Color to Blue
lut = np.zeros((256, 1, 3), dtype = 'uint8')
lut[:, 0, 2] = np.arange(256)

# Applying the LUT on the image
blueImg = cv.LUT(img, lut)

# Plotting the Original image and
# the image with Changed Color
plt.figure(figsize=(15,8))
plt.subplot(1, 2, 1)
plt.imshow(img)
plt.title('Original Image')

plt.subplot(1, 2, 2)
plt.imshow(blueImg)
plt.title('Color Changed to Blue')
plt.show()
```

```python
# Creating a LUT to Change the Image Color to Green
lut = np.zeros((256, 1, 3), dtype = 'uint8')
lut[:, 0, 1] = np.arange(256)

# Applying the LUT on the image
greenImg = cv.LUT(img, lut)

# Plotting the Original image and
# the image with Changed Color
plt.figure(figsize=(15,8))
plt.subplot(1, 2, 1)
plt.imshow(img)
plt.title('Original Image')

plt.subplot(1, 2, 2)
plt.imshow(greenImg)
plt.title('Color Changed to Green')
plt.show()
```
