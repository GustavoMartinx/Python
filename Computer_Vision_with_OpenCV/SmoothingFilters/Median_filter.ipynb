## Median Filter

This non-linear filter classified as a low-pass filter presents good results in the treatment of images with the type of noise called salt and pepper. In this sense, compared to Gaussian filter, its is more effective in preserving high frequencies details (such as image borders and contours). <br><br>

Mathematically, this filter acts by changing the value of each target pixel by the statistical median of adjacent pixel values.

##### Library Importing

```python
import cv2 as cv
import matplotlib.pyplot as plt
%matplotlib inline
```

##### Image Reading

```python
img = cv.imread('./Resources/Img/head.png')
```

`medianBlur(img, smoothing)`

* Input: <br>
    1.  `img`: Image to which the mask will be applied; <br>
    2.  `smoothing`: Degree of smoothing (unique, odd and integer). <br><br>

* Output: Matrix referring to the filtered image.

```python
medianBlurredImg = cv.medianBlur(img, 5)
```

```python
# Image Plotting
fig = plt.figure(figsize=(20, 50))

ax1 = fig.add_subplot(121)
plt.imshow(img)
plt.title("Original Image - Salt and Pepper Noise")

ax1 = fig.add_subplot(122)
plt.imshow(medianBlurredImg)
plt.title("Image with Median Filter Applied")
plt.show()
```
