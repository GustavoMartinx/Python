## Gaussian Filter

This linear filter classified as a low-pass filter and presents good results in the treatment of images with Gaussian noise. In this sense, as with the Average Filter, the result is an image with its high frequencies suppressed.

##### Library Importing

```python
import cv2 as cv
import matplotlib.pyplot as plt
%matplotlib inline
```

##### Image Reading

```python
img = cv.imread('./Resources/Img/bacteria.png')
```

`GaussianBlur(img, mask, smoothing)`

* Input: <br>
    1.  `img`: Image to which the mask will be applied; <br>
    2.  `mask`: Mask dimension that will be applied to the image. <br>
    3.  `smoothing`: degree of smoothing. <br><br>

* Output: Matrix referring to the image that had its high frequencies suppressed.

```python
GaussianBlurredImg = cv.GaussianBlur(img, (5, 5), 0)
```

```python
# Image Plotting
fig = plt.figure(figsize=(20, 50))

ax1 = fig.add_subplot(121)
plt.imshow(img)
plt.title("Original Image - Gaussian Noise")

ax1 = fig.add_subplot(122)
plt.imshow(GaussianBlurredImg)
plt.title("Image with Gaussian Filter Applied")
plt.show()
```
