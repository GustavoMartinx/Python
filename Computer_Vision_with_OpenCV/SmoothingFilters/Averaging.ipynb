## Averaging Filter

This linear filter classified as a low-pass filter replace each pixel in the image by the average value of its adjacent ones

##### Library Importing

```python
import cv2 as cv
import matplotlib.pyplot as plt
%matplotlib inline
```

##### Image Reading

```python
img = cv.imread('./Resources/Img/einstein.png')
```

`blur(img, mask)`

* Input: <br>
    1.  `img`: Image to which the mask will be applied; <br>
    2.  `mask`: Mask that will be applied to the image. <br><br>

* Output: Matrix referring to the treated image.

```python
imgMatrix = cv.blur(img, (5, 5))

# Image Plotting
fig = plt.figure(figsize=(20, 50))

ax1 = fig.add_subplot(121)
plt.imshow(img)
plt.title("Original Image - Noise")

ax1 = fig.add_subplot(122)
plt.imshow(imgMatrix)
plt.title("Filtered Image")
plt.show()
```
