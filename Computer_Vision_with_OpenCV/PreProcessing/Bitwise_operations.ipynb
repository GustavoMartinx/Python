## Bitwise Operations

This type of operation can be used to create image masks or watermarks, for example, resulting in a new image. Furthermore, such operations are applied to each bit of the image separately for more precision. Furthermore, this resource is used to extract any part of the image (to work with non-rectangular ROIs).

##### Library Importing

```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

##### Image Creation and Plotting

```python
# Creation of Image 1: Rectangle
img1 = np.zeros((300, 300), dtype="uint8") # creates a 300x300 matrix
# creates a rectangle starting at (100, 100) point and ending at (250, 250) point
# with a backgroud black and its white (-1 param)
cv.rectangle(img1, (100, 100), (250, 250), 255, -1)

# Creation of Image 2: Circle
img2 = np.zeros((300, 300), dtype="uint8")
cv.circle(img2, (150, 150), 90, 255, -1) # Radius value: 90

# Image Plotting
plt.figure(figsize=(10, 8))
plt.subplot(1, 2, 1)
plt.imshow(img1, cmap='gray')
plt.subplot(1, 2, 2)
plt.imshow(img2, cmap='gray')
plt.show()
```

#### AND Operation

Calculates the conjunction of pixels that are common in both images and the remaining pixels are removed from the output image.

```python
rectangle_and_circle = cv.bitwise_and(img1, img2)

plt.figure(figsize=(10, 8))
plt.imshow(rectangle_and_circle, cmap='gray')
plt.show()
```

#### OR Operation

Calculate the disjunction of pixels in both images, that is, only the pixels that are in one image and not in the other are considered.

```python
rectangle_or_circle = cv.bitwise_or(img1, img2)

plt.figure(figsize=(10, 8))
plt.imshow(rectangle_or_circle, cmap='gray')
plt.show()
```

#### XOR Operation

Inverts the pixels that cross images 1 and 2 and the remaining pixels remain the same.

```python
rectangle_xor_circle = cv.bitwise_xor(img1, img2)

plt.figure(figsize=(10, 8))
plt.imshow(rectangle_xor_circle, cmap='gray')
plt.show()
```

#### NOT Operation

Inverts each pixel belonging to the image (replaces white pixels with black ones and vice versa). Note: This operation can be performed on only one image at a time.

```python
rectangle_not_circle = cv.bitwise_not(img1)

plt.figure(figsize=(10, 8))
plt.imshow(rectangle_not_circle, cmap='gray')
plt.show()
```
